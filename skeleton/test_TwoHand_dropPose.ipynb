```python
# DataGenerator
class DataGenerator(Sequence):
    def __init__(self, x_set, y_set, batch_size):
        self.x, self.y = x_set, y_set
        self.batch_size = batch_size

    def __len__(self):
        return int(np.ceil(len(self.x) / float(self.batch_size)))

    def __getitem__(self, idx):
        batch_x = self.x[idx * self.batch_size:(idx + 1) * self.batch_size]
        batch_y = self.y[idx * self.batch_size:(idx + 1) * self.batch_size]
        return batch_x, batch_y

train_gen = DataGenerator(x_train, y_train_1h, 16)
test_gen = DataGenerator(x_train, y_test_1h, 16)
```

```python
check_group = df_drop_ax0.groupby('class').size()
check_group[:]
```

```python
def LSTM_model_TwoHand():
    model = Sequential()

    # model.add(LSTM(units = 32, input_shape=(step, 1), activation= 'tanh'))

    model.add(LSTM(units = 256, input_shape=(step, 1), activation= 'tanh', return_sequences= True))
    model.add(LSTM(units = 128, activation='tanh', return_sequences=True))
    model.add(LSTM(units= 128, activation= 'tanh'))

    # model.add(Flatten())
    model.add(Dense(units = 128, activation = 'tanh'))
    #model.add(Dropout(0.2))
    model.add(Dense(units = 84, activation = 'tanh'))
    #model.add(Dropout(0.2))
    model.add(Dense(units= 16, activation= 'softmax'))
    model.compile(optimizer= 'adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

model = LSTM_model_TwoHand()
model.build()
model.summary()
```

```python
es = EarlyStopping(monitor='val_loss', verbose=1, patience=7) #
mc = ModelCheckpoint('TL1 tr300.h5', monitor='val_accuracy', verbose=1, save_best_only=True) #

history = model.fit(x_train,y_train_1h,epochs=100,batch_size=16,verbose=1,validation_split=0.2, callbacks=[es, mc])
# My_model = model.fit(train_gen, epochs=50, verbose=1, validation_data=test_gen, callbacks=[es, mc])
```

```python
df_hist = pd.DataFrame.from_dict(history.history) # draw graph
```

```python
print("Training Done")

df_hist['loss'].plot(style='b--',label='train')
df_hist['val_loss'].plot(style='r-',label='val')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
plt.title('Loss')
plt.show()
```

```python
df_hist['accuracy'].plot(style='b--',label='train')  # 68, 16
df_hist['val_accuracy'].plot(style='r-',label='val')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend()
plt.title('Accuracy')
plt.show()
```

```python
score = model.evaluate(x_test,y_test_1h,verbose=0) # test = 36
print('test loss : ',score[0])
print('test accuracy : ',score[1])
```
