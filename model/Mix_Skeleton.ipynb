```python
# DataGenerator
class DataGenerator(Sequence):
    def __init__(self, x_set, y_set, batch_size):
        self.x, self.y = x_set, y_set
        self.batch_size = batch_size

    def __len__(self):
        return int(np.ceil(len(self.x) / float(self.batch_size)))

    def __getitem__(self, idx):
        batch_x = self.x[idx * self.batch_size:(idx + 1) * self.batch_size]
        batch_y = self.y[idx * self.batch_size:(idx + 1) * self.batch_size]
        return batch_x, batch_y

one_train_gen = DataGenerator(one_x_train, one_y_train_1h, 16)
one_test_gen = DataGenerator(one_x_train, one_y_test_1h, 16)

two_train_gen = DataGenerator(two_x_train, two_y_train_1h, 16)
two_test_gen = DataGenerator(two_x_train, two_y_test_1h, 16)
```

```python
def LSTM_model_OneHand():
    one_model = Sequential()

    # model.add(LSTM(units = 32, input_shape=(step, 1), activation= 'tanh'))

    one_model.add(LSTM(units = 256, input_shape=(one_step, 1), activation= 'tanh', return_sequences= True))
    one_model.add(LSTM(units = 128, activation='tanh', return_sequences=True))
    one_model.add(LSTM(units= 128, activation= 'tanh'))

    # model.add(Flatten())
    one_model.add(Dense(units = 128, activation = 'tanh'))
    #model.add(Dropout(0.2))
    one_model.add(Dense(units = 84, activation = 'tanh'))
    #model.add(Dropout(0.2))
    one_model.add(Dense(units= 73, activation= 'softmax'))
    one_model.compile(optimizer= 'adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return one_model

one_model = LSTM_model_OneHand()
one_model.build()
one_model.summary()
```

```python
def LSTM_model_TwoHand():
    two_model = Sequential()

    # model.add(LSTM(units = 32, input_shape=(step, 1), activation= 'tanh'))

    two_model.add(LSTM(units = 256, input_shape=(two_step, 1), activation= 'tanh', return_sequences= True))
    two_model.add(LSTM(units = 128, activation='tanh', return_sequences=True))
    two_model.add(LSTM(units= 128, activation= 'tanh'))

    # model.add(Flatten())
    two_model.add(Dense(units = 128, activation = 'tanh'))
    #model.add(Dropout(0.2))
    two_model.add(Dense(units = 84, activation = 'tanh'))
    #model.add(Dropout(0.2))
    two_model.add(Dense(units= 16, activation= 'softmax'))
    two_model.compile(optimizer= 'adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return two_model

two_model = LSTM_model_TwoHand()
two_model.build()
two_model.summary()
```

```python
one_es = EarlyStopping(monitor='val_loss', verbose=1, patience=7) #
one_mc = ModelCheckpoint('OneHand.h5', monitor='val_accuracy', verbose=1, save_best_only=True) #

one_history = one_model.fit(one_x_train,one_y_train_1h,epochs=4,batch_size=16,verbose=1,validation_split=0.2, callbacks=[one_es, one_mc])
# My_model = model.fit(train_gen, epochs=50, verbose=1, validation_data=test_gen, callbacks=[es, mc])
```
