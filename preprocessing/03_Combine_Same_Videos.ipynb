```python
import os
import shutil
```

```python
src_folder = 'D:\\FingerSpelling\\TFS_Dataset\\01_RawData\\'
desPath = 'D:\\FingerSpelling\\TFS_Dataset\\02_Combine_Data\\'
allFrame = 0
countDir = 0
countClips = 0

num_name = 0
num_sub = 0
num_mainfile = 0
num_file = 0
folder_missing = 0
file_missing = 0
```

```python
for name_folder in os.listdir(src_folder):
    if name_folder[0] == 'S':
        num_name += 1
        name_path = src_folder + name_folder + '\\'
        # print('num_name = ',num_name)
        countDir = countDir + 1
        print('Signer : ',countDir)
        for subtype_folder in os.listdir(name_path):
            num_sub += 1
            subtype_path = name_path + subtype_folder + '\\'
            # print('num_sub = ', num_sub)
            for mainfile_folder in os.listdir(subtype_path):
                num_mainfile += 1
                # print('Main File Folder: ',num_mainfile)
                mainfile_path = subtype_path + mainfile_folder + '\\'
                # print('mainfile_path = ',mainfile_path)
                # print('num_mainfile = ',num_mainfile)
                for file in os.listdir(mainfile_path):
                    try:
                        num_file += 1
                        srcfile = mainfile_path + file

                        print('srcfile = ', srcfile)
                        desfile = desPath + subtype_folder + '\\'  + mainfile_folder + '\\'
                        print('desfile = ', desfile)
                        countClips += 1
                        shutil.copy(srcfile, desfile)
                    except FileExistsError:
                        print("File " , file ,  " already exists")
print('')
print('All clips = ', countClips)
print('Done !!!')
```
